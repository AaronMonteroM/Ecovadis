```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import numpy as np
from sklearn import preprocessing
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pacmap
from sklearn.metrics import silhouette_score

df=pd.read_excel('../data/raw/Assignment (churn prediction).xlsx')

#Plot function

def plot_distrib_num_feature(feat,bins):
    plt.hist(df[feat],bins)
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()   
    return None

def plot_distrib_target_num_feature(feat,bins):
    plt.hist(df[feat].loc[df['Exited']==0],bins)
    plt.hist(df[feat].loc[df['Exited']==1],bins)
    plt.legend(['0','1'],title='Exited')
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()

    return None

def plot_distrib_cat_feature(feat):
    df[feat].value_counts().sort_index().plot(kind='bar', rot=0)
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()    
    return None

def plot_distrib_target_cat_feature(feat):
    df[feat].loc[df['Exited']==0].value_counts().sort_index().plot(kind='bar', rot=0, position=1, width=0.25)
    df[feat].loc[df['Exited']==1].value_counts().sort_index().plot(kind='bar', color='red', rot=0, position=0, width=0.25)
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()  
    return None

def plot_distrib_few_num_feature(feat):
    df[feat].astype(int).value_counts().sort_index().plot(kind='bar', rot=0)
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()    
    return None

def plot_distrib_target_few_num_feature(feat):
    df[feat].loc[df['Exited']==0].astype(int).value_counts().sort_index().plot(kind='bar', rot=0, position=1, width=0.25)
    df[feat].loc[df['Exited']==1].astype(int).value_counts().sort_index().plot(kind='bar', color='red', rot=0, position=0, width=0.25)
    plt.xlabel(feat)
    plt.ylabel('# Customers')
    plt.show()  
    return None
```

## CustomerFeedback

We will classify the feedback comments provided by the customers as positive, neutral and negative using the nltk library for natural language processing.

```python
nltk.downloader.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
sent_comment=df['CustomerFeedback'].loc[df['CustomerFeedback'].isna()==False].apply(lambda x: sia.polarity_scores(x)["compound"])
sent_comment_lvl=np.select([sent_comment < 0, sent_comment == 0, sent_comment > 0],
                           [-1,0,1])
```

```python
values, counts = np.unique(sent_comment_lvl, return_counts=True)
print(values)
print(counts)
```

Of all the comments registered by customers, only 4 have been classified as neutral. Since, as we saw previously, we have 70% missing values for this feature, we are going to consider that those customers who did not register any comments have a neutral opinion about our company.

```python
sent_comment_col=np.zeros(len(df))
sent_comment_col[sent_comment.index]=sent_comment_lvl
df['CustomerFeedback']=sent_comment_col
```

Therefore, the total distribution of customer feedback is:

```python
df['CustomerFeedback'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.xlabel('CustomerFeedback')
plt.ylabel('# Customers')
plt.xticks(np.arange(3),['negative','neutral','positive'])
plt.show()
```

If we visualize the distribution of customer feedback based on the target class, we see, as expected, that those with a negative opinion of our company are more likely to leave.

```python
df['CustomerFeedback'].loc[df['Exited']==0].value_counts().sort_index().plot(kind='bar', rot=0, position=1, width=0.25)
df['CustomerFeedback'].loc[df['Exited']==1].value_counts().sort_index().plot(kind='bar', color='red', rot=0, position=0, width=0.25)
plt.xticks(np.arange(3),['negative','neutral','positive'])
plt.xlabel('CustomerFeedback')
plt.ylabel('# Customers')
plt.show()
```

# Data transformation

Once we have seen the features separately, we are going to analyze them as a whole. First, we will observe the correlation and later we will spatially visualize our customers (applying dimensionality reduction methods), to observe the cohesion of the target class in order to apply an appropriate data balancing method. To visualize the customers, it is necessary to pass the categorical values to numeric ones and perform a standardization or normalization of the data.

## Categorical values encoding

After removing the Surname column and transforming CustomerFeedback, we are left with only 2 categorical features: Country and Gender.

```python
le_country = preprocessing.LabelEncoder()
df['Country']=le_country.fit_transform(df['Country'])
le_gender = preprocessing.LabelEncoder()
df['Gender']=le_gender.fit_transform(df['Gender'])
df
```

## Data standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_stand=df.copy()
df_stand.iloc[:,:-1]=scaler.fit_transform(df_stand.iloc[:,:-1])
df_stand
```

# Correlation matrix

```python
corr = df_stand.corr().round(2)
sns.heatmap(corr, annot=True)
```

As we can see, there is not a great correlation between the features. Therefore, we cannot eliminate redundant features (that convey similar information), nor select only those features that have a high correlation with the target class.

# Customer Visualization

First of all we are going to try the classical method for dimensionality reduction, PCA. To analyze if this method is useful to perform the reduction of our data (excluding the target class), we will force the transformation to do it while maintaining an explained variance of 95% and we will observe in how many dimensions it reduces the 11 initial features.

```python
pca = PCA(n_components = 0.95)
X = pca.fit_transform(df_stand.iloc[:,:-1])
print(X.shape)
```

Unfortunately PCA is not able to reduce the dimensionality of our data without losing a considerable amount of information. Therefore, we are going to apply a non-linear dimensionality reduction method, such as PaCMAP. This method allows us to preserve the local and global structure of the original data, although we will not know how much information we will really be losing with this transformation (as happens with all non-linear methods such as t-SNE or UMAP).

```python
embedding = pacmap.PaCMAP() 
X_transformed = embedding.fit_transform(np.array(df_stand.iloc[:,:-1]))
y=np.array(df_stand.iloc[:,-1])

s=plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y, s=2)
plt.legend(handles=s.legend_elements()[0], labels=['0','1'],title='Exited')
plt.show()
```

As we can see, the customers who stay and leave our company are intermingled with each other. Because of this distribution, random oversampling might work better than creating synthetic data through methods like SMOTE.

To confirm that this lack of cohesion is not produced solely by dimensionality reduction, we will calculate the Silhouette coefficient for the original and transformed data. This coefficient will tell us how cohesive a class is and how separated it is from the other. It is a method oriented to clustering problems, but in this case it is useful for evaluating the target class.

```python
print('Silhouette coefficient original data:'+str(silhouette_score(X, y)))
print('Silhouette coefficient transformed data:'+str(silhouette_score(X_transformed, y)))
```

As we can see, although the transformation would make us lose information, the Silhouette coefficient is similar and low (poor cohesion and separability) in both cases.
